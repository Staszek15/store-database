```python
def generate_results():
    tournaments = pd.read_csv('tournaments.csv')
    customers = pd.read_csv('customers.csv')
    tournament_id = tournaments.tournament_id
    
    participants_list = []
    score_list = []
    for i in range(len(tournaments)):
        row = tournaments.iloc[i]

    if row.team_players_number > 1:
        pass
    else:
        customers_id = random.sample(sorted(customers.customer_id), row.total_players_number)
        score = np.arange(1,len(customers_id) + 1)
        
        participants_list.extend(customers_id)
        score_list.extend(score)
        
    return()
        
    
    
```

```python
generate_results()
```
